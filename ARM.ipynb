### 1. Data preprocessing

```python
import pandas as pd

snu_df = pd.read_csv('./crowling/snuh_crowling.csv')
asan_df = pd.read_csv('./crowling/asan_crowling.csv')

print(list(snu_df["질환명"]))
print(list(asan_df["질환명"]))
```

```python
for i in range(len(asan_df["동의어"])):
    syn_list = asan_df["동의어"][i].split(",")
    new_syns = []
    for syn in syn_list:
        syn = syn.replace("[", "").replace("]", "").replace("'", "")
        if syn.replace(" ","").encode().isalpha() == False:
            syn = syn.replace(" ", "")
        new_syns.append(syn)
    asan_df.loc[i, "동의어"] = new_syns
asan_df.head()
```

```python
df = pd.DataFrame(columns=["질환명", "동의어", "증상 키워드", "증상1", "증상2", "URL1", "URL2"])

for i in range(len(snu_df)):
    if snu_df["질환명"][i].replace(" ", "") != asan_df["질환명"][i].replace(" ", ""):
        
        # print(asan_df["동의어"][i])
        if snu_df["질환명"][i].replace(" ", "") in asan_df["동의어"][i]:
            df.loc[i, "질환명"] = snu_df["질환명"][i]
        else:
            df.loc[i, "질환명"] = snu_df["질환명"][i]
            # syns_list = asan_df["동의어"][i].deepcopy()
            asan_df["동의어"][i].insert(0, asan_df["질환명"][i].replace(" ", ""))
            print(f"{snu_df['질환명'][i]}의 정보가 일치하지 않습니다. 다시 확인하세요.")
    else:
        df.loc[i, "질환명"] = snu_df["질환명"][i]

    df.loc[i, '동의어'] = asan_df["동의어"][i]
    df.loc[i, '증상 키워드'] = snu_df["증상 키워드"][i] + "," + asan_df["증상 키워드"][i]
    df.loc[i, '증상1'] = snu_df["증상"][i]
    df.loc[i, '증상2'] = asan_df["증상"][i]
    df.loc[i, 'URL1'] = snu_df["URL"][i]
    df.loc[i, 'URL2'] = asan_df["URL"][i]
    
df
```

```python
for i in range(len(df['증상 키워드'])):
    syts = df['증상 키워드'][i].split(",")
    # print(syts)
    new_syts = []
    for syt in syts:
        syt = syt.replace(" ", "")
        
        if syt != "":
            if syt == "발열":
                new_syts.append("열")
            elif syt == "복부통증":
                new_syts.append("복통")
            elif syt == "호흡장애":
                new_syts.append("호흡곤란")
            elif syt == "어지러움":
                new_syts.append("어지럼증")
            elif syt == "비폐색":
                new_syts.append("코막힘")
            else:
                new_syts.append(syt)
    df.loc[i, '증상 키워드'] = list(set(new_syts))
    
# df.to_csv('./crowling/merged_crowling.csv', index=False, encoding='utf-8-sig')
df
```

```python
import itertools

all_subsets = []
for i in range(len(user_symptoms)):
    all_subsets.extend(list(list(itertools.combinations(user_symptoms, i+1))))
all_subsets
```

```python
keyword_combi = pd.DataFrame(columns=["keyword", "support"])

for i in range(len(symptoms_sets)):
    for subset in all_subsets:
        if set(subset) == (set(symptoms_sets['itemsets'][i])):
            keyword_combi.loc[len(keyword_combi)] = \
                [list(symptoms_sets['itemsets'][i]), symptoms_sets['support'][i]]
keyword_combi.sort_values(['support'], ascending=False)
```

```python
user_syt_df = pd.DataFrame(columns=rules.columns)
for i in range(len(keyword_combi)):        
    filtered = rules[rules["antecedents"] == frozenset(keyword_combi["keyword"][i])]
    filtered.sort_values(['support'], ascending=False)
    user_syt_df = pd.concat([user_syt_df, filtered])
user_syt_df.sort_values(['confidence'], ascending=False)
```

```python
item_sets = []
for i in range(len(user_syt_df)):
    item_sets.append(list(user_syt_df['antecedents'].values[i]) + list(user_syt_df['consequents'].values[i]))
item_sets
```

```python
best_matchs = []
able_diseases = []
able_df_disases = []

for i in range(len(df["증상 키워드"])):
    if set(user_symptoms) == set(df["증상 키워드"][i]):
        best_matchs.append(df["질환명"][i])
        print("best")
    else:
        if len(item_sets) != 0:        
            for item_set in item_sets:
                if set(item_set).issubset(set(df["증상 키워드"][i])):
                    able_diseases.append(df["질환명"][i])
        else:
            if set(user_symptoms).issubset(set(df["증상 키워드"][i])):
                able_df_disases.append(df["질환명"][i])
        
best_matchs = list(set(best_matchs))
able_diseases = list(set(able_diseases))
able_df_disases = list(set(able_df_disases))

if len(best_matchs) != 0:
    print(f"ARM 기반 입력하신 증상에 대해 가장 적합한 질환은 {best_matchs}입니다.")
if len(able_diseases) != 0:
    print(f"ARM 기반 입력하신 증상에 대해 가능성 있는 질환은 {able_diseases}입니다.")
if len(able_df_disases) != 0:
    print(f"Dataframe 내 탐색 기반 입력하신 증상에 대해 가능성 있는 질환은 {able_df_disases}입니다.")
```
