```python
import pandas as pd

snu_df = pd.read_csv('./crowling/snuh_crowling_ver_2.csv')
asan_df = pd.read_csv('./crowling/asan_crowling_ver_2.csv')

print(list(snu_df["질환명"]))
print(list(asan_df["질환명"]))
```

```python
for i in range(len(asan_df["동의어"])):
    syn_list = asan_df["동의어"][i].split(",")
    new_syns = []
    for syn in syn_list:
        syn = syn.replace("[", "").replace("]", "").replace("'", "")
        if syn.replace(" ","").encode().isalpha() == False:
            syn = syn.replace(" ", "")
        new_syns.append(syn)
    asan_df.loc[i, "동의어"] = new_syns
asan_df.head()
```

```python
print(list(snu_df.columns))
print(list(asan_df.columns))
```

```python
df = pd.DataFrame(columns=["질환명", "동의어", "증상 키워드", "증상1", "증상2", "치료1", "치료2", "경과1", "경과2", "생활가이드", "주의사항", "예방방법", "FAQ", "URL1", "URL2"])

for i in range(len(snu_df)):
    if snu_df["질환명"][i].replace(" ", "") != asan_df["질환명"][i].replace(" ", ""):
        
        # print(asan_df["동의어"][i])
        if snu_df["질환명"][i].replace(" ", "") in asan_df["동의어"][i]:
            df.loc[i, "질환명"] = snu_df["질환명"][i]
        else:
            df.loc[i, "질환명"] = snu_df["질환명"][i]
            # syns_list = asan_df["동의어"][i].deepcopy()
            asan_df["동의어"][i].insert(0, asan_df["질환명"][i].replace(" ", ""))
            print(f"{snu_df['질환명'][i]}의 정보가 일치하지 않습니다. 다시 확인하세요.")
    else:
        df.loc[i, "질환명"] = snu_df["질환명"][i]

    df.loc[i, '동의어'] = asan_df["동의어"][i]
    df.loc[i, '증상 키워드'] = snu_df["증상 키워드"][i] + "," + asan_df["증상 키워드"][i]
    
    df.loc[i, '증상1'] = snu_df["증상"][i]
    df.loc[i, '증상2'] = asan_df["증상"][i]
    df.loc[i, '치료1'] = snu_df["치료"][i]
    df.loc[i, '치료2'] = asan_df["치료"][i]
    df.loc[i, '경과1'] = snu_df["경과/합병증"][i]
    df.loc[i, '경과2'] = asan_df["경과"][i]
    df.loc[i, 'URL1'] = snu_df["URL"][i]
    df.loc[i, 'URL2'] = asan_df["URL"][i]
    
    df.loc[i, '생활가이드'] = snu_df["생활가이드"][i]
    df.loc[i, '예방방법'] = snu_df["예방방법"][i]
    df.loc[i, '주의사항'] = asan_df["주의사항"][i]
    df.loc[i, 'FAQ'] = asan_df["FAQ"][i]
    
df.head()
```

```python
df.insert(3, "증상 등장 빈도", None)
symptoms = []

for i in range(len(df['증상 키워드'])):
    syts = df['증상 키워드'][i].split(",")
    # print(syts)
    new_syts = []
    for syt in syts:
        syt = syt.replace(" ", "")
        
        if syt != "":
            if syt == "발열":
                new_syts.append("열")
            elif syt == "복부통증":
                new_syts.append("복통")
            elif syt == "호흡장애":
                new_syts.append("호흡곤란")
            elif syt == "어지러움":
                new_syts.append("어지럼증")
            elif syt == "비폐색":
                new_syts.append("코막힘")
            else:
                new_syts.append(syt)
    df.at[i, '증상 키워드'] = list(set(new_syts))
    symptoms.append(list(set(new_syts)))
    df.at[i, '증상 등장 빈도'] = new_syts
    
df.head()
```

```python
for i in range(len(df)):
    keys = df["증상 등장 빈도"][i]
    temp_dict = dict.fromkeys(keys, 0)
    for key in keys:
        temp_dict[key] += 1
    print(temp_dict)
    # raise
    df.loc[i, "증상 등장 빈도"] = [temp_dict]
```

```python
df.fillna('.', inplace=True)
```

```python
cols = ['생활가이드', '주의사항', '예방방법', 'FAQ']
for i in range(len(df)):
    for col in cols:
        df.at[i, col] = df[col][i].replace("한다.", "합니다.").replace("하다.", "합니다.").replace("이다.", "입니다.").replace("된다.", "됩니다.").replace("는다.", "습니다.") \
            .replace("있다.", "있습니다.")
        for j in range(len(df[col][i])):
            if df[col][i][j:j+2] == "다.":
                if df[col][i][j-1] not in ["한", "하", "이", "니", "된", "는", "있"]:
                    df[col][i] = df[col][i][:j] + "습니다." + df[col][i][j+2:]
        print("=======")            
        print(df[col][i])
```

```python
df.to_csv('./crowling/merged_crowling_ver_2.csv', index=False, encoding='utf-8-sig')
```
